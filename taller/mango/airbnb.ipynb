```python
import numpy as np
import pandas as pd

from dateutil import parser
from dateutil.relativedelta import relativedelta

from datetime import timedelta
from datetime import datetime

import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = [10,6]
mpl.rcParams['figure.dpi'] = 80
mpl.rcParams['savefig.dpi'] = 200
mpl.rcParams['font.size'] = 17
mpl.rcParams['legend.fontsize'] = 'large'
mpl.rcParams['figure.titlesize'] = 'medium'
mpl.rcParams['lines.linewidth'] = 2.5
mpl.rcParams['lines.markersize'] = 10
```

```python
# Load file
df_airbnb = pd.read_csv('data/2018-01.csv', sep=',')
df_airbnb.head()
```

```python
start_date = "2018-04-01"
stop_date = "2018-09-01"

start = datetime.strptime(start_date, "%Y-%m-%d")
stop = datetime.strptime(stop_date, "%Y-%m-%d")

path = 'data/'+start.strftime('%Y-%m')+'.csv'
df_airbnb = pd.read_csv(path, sep=',')
last_month_ids = df_airbnb.id.values

while start < stop:
    start = start + relativedelta(months=+1)
    path = 'data/'+start.strftime('%Y-%m')+'.csv'
    df_airbnb = pd.read_csv(path, sep=',')
    current_month_ids = df_airbnb.id.values
    print start.strftime('%Y-%m')
    print 'last month', len(last_month_ids)
    print 'current month', len(current_month_ids)
    print 'new listings', len(current_month_ids) - len(np.intersect1d(last_month_ids, df_airbnb.id.values))
    print 'erased listings', len(last_month_ids) - len(np.intersect1d(last_month_ids, df_airbnb.id.values))
    print '\n'
    
    last_month_ids = current_month_ids
    
```
